```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math, Markdown
```

```python
k = 1
l = 10
l1 = 5
m = 100
g = 9.8
B = 3

w1 = np.sqrt(g / l - B**2 / (4 * m**2))
w2 = np.sqrt(2 * k / m * (l1 / l) + g / l - B**2 / (4 * m**2))

p0 = 10

display(Math(r'ω_1 = {}'.format(w1)))
display(Math(r'ω_2 = {}'.format(w2)))
```

```python
def phi1(t):
    return p0 * np.exp(-B / (2 * m) * t) * np.cos((w1 + w2)/2 * t) * np.cos((w1 - w2)/2 * t)

def phi2(t):
    return p0 * np.exp(-B / (2 * m) * t) * np.sin((w1 + w2)/2 * t) * np.sin((w2 - w1)/2 * t)

def v_phi1(t):
    return -B / (4 * m) * p0 * np.exp(-B / (2 * m) * t) * (np.cos(w1 * t) + np.cos(w2 * t)) - 1/2 * p0 * np.exp(-B / (2 * m) * t) * (w1 * np.sin(w1 * t) + w2 * np.sin(w2 * t))

def v_phi2(t):
    return -B / (4 * m) * p0 * np.exp(-B / (2 * m) * t) * (np.cos(w1 * t) - np.cos(w2 * t)) - 1/2 * p0 * np.exp(-B / (2 * m) * t) * (w1 * np.sin(w1 * t) - w2 * np.sin(w2 * t))
```

```python
t0 = 0.0 
tf = 300.0  
dt = 0.01

num_steps = int((tf - t0) / dt) + 1

theta1 = np.zeros(num_steps)
theta2 = np.zeros(num_steps)
v_theta1 = np.zeros(num_steps)
v_theta2 = np.zeros(num_steps)
time = np.linspace(t0, tf, num_steps)

for i in range(num_steps):
    theta1[i] = phi1(t0 + i*dt)
    theta2[i] = phi2(t0 + i*dt)
    v_theta1[i] = v_phi1(t0 + i*dt)
    v_theta2[i] = v_phi2(t0 + i*dt)
```

```python
plt.figure(figsize=(15, 10))
plt.plot(time, theta1, label='Pendulum 1')
plt.plot(time, theta2, label='Pendulum 2')
plt.title('Coupled Pendulum Motion')
plt.xlabel('Time (s)')
plt.ylabel('Angle (radians)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.plot(time, v_theta1, label='Pendulum 1')
plt.plot(time, v_theta2, label='Pendulum 2')
plt.title('Coupled Pendulum Motion')
plt.xlabel('Time (s)')
plt.ylabel('Velocity (radians/s)')
plt.legend()
plt.grid(True)
plt.show()
```
